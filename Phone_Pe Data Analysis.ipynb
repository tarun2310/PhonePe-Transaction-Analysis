```python
#Task 3:
#Data Quality Checks
#Ensure Data Consistency across District and State level

district_data=district_txn_user_data.groupby('State').agg({'Transactions':'sum','Amount (INR)':'sum','Registered Users':'sum'})
print(district_data.shape)
#display(district_data)

state_data=state_txn_user_data.groupby('State').agg({'Transactions':'sum','Amount (INR)':'sum','Registered Users':'sum'})
print(state_data.shape)
#display(state_data)

merged_data=district_data.merge(state_data,on='State',how='inner',suffixes=('_state', '_district'))
merged_data['Transactions_Match']=merged_data['Transactions_state']==merged_data['Transactions_district']

merged_data['Amount_Match']=merged_data['Amount (INR)_state']==merged_data['Amount (INR)_district']


merged_data['Users_Match']=merged_data['Registered Users_state']==merged_data['Registered Users_state']

merged_data
```

```python
# Task 4
# Data Merging and Advanced Analysis
# Ratio of Users to population by state
# 1. Merge the State Txn and user dataset with the district demographics dataset to calculate the ratio of Registered users to the 
# population of each state. Display the results in a tabular format.

display(state_txn_user_data.head())
display(district_demographics_data.head())
```

```python
#filtering out the latest data for states in state Txn dataset


#finding out total population
district_total_Population=district_demographics_data.groupby('State')['Population'].sum().reset_index()
state_txn_user_data_merged=state_txn_user_data.groupby('State')['Registered Users'].sum().reset_index()
state_district_merged=state_txn_user_data_merged.merge(district_total_Population,on='State',how='inner')

#Rounded it to 2 decimal places for better readability
state_district_merged['UsersToPopulation']= round(state_district_merged['Registered Users']/state_district_merged['Population'],2)
state_district_merged=state_district_merged.sort_values(by='UsersToPopulation',ascending=True)
#Taking High users states to plot chart
#Highest_users_data=state_district_merged.sort_values(by='UsersToPopulation',ascending=False).reset_index(drop=True).head()
#Lowest_users_data=state_district_merged.sort_values(by='UsersToPopulation',ascending=False).reset_index(drop=True).tail()
#Highest_users_data
```

```python
#Chart for State with Users

plt.figure(figsize=(12,8))
plt.barh(state_district_merged['State'],state_district_merged['UsersToPopulation'],color='skyblue')
plt.xlabel('State')
plt.ylabel('Ratio')
plt.title('Registered Users to Population Ratio')
plt.tight_layout()
plt.show()
```

### Task 4.2 Correlate Population Density with Transaction Volume

#### Merge the District Txn and Users Dataset with the District Demographics Dataset

### Calculate the Correlation between Population Density and Transaction Volume

```python
corr_pop_Tran=district_demo_merged_data['Density'].corr(district_demo_merged_data['Transactions'])

print(f'Correlation between Population Density and Transaction Volume {corr_pop_Tran:.2f}')
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(data=district_demo_merged_data,x='Density',y='Transactions')

# regression line
sns.regplot(x='Density', y='Transactions', data=district_demo_merged_data, scatter=False, color='red', ci=None)

plt.title(f'Scatter Plot with Regression Line\nCorrelation: {district_demo_merged_data["Density"].corr(district_demo_merged_data["Transactions"]):.2f}')
plt.tight_layout()
plt.show()
```

### Insights

* User activity (Transactions & Amount) is strongly driven by the number of Registered Users.
* Population and Density have only modest influence, suggesting that raw population doesn't ensure high digital payment activity.
* Increasing registered users could significantly boost digital payment transactions and amounts.
* Xiaomi is the most Popular Brand in 35 out of 36 State/Union Territories.
* Ladakh has the highest Average Transaction Value- 3408.222984 whereas West Bengal has the lowest Average Transaction Value     1373.600590.
* States with the high Population Density tends to have more transactions, but not always.(0.31 correlation does not indicate strong Correlation.
* In Most of the states, App is used mainly for Peer-to-peer Transaction followed by Merchant Payments and Recharge and Bill Payments(for example=Madhya Pradesh)
* Transactions have increased significantly from 2018 to 2021 in all States.
